```python
from idlelib.debugger_r import restart_subprocess_debugger
from tabnanny import verbose

from crewai.project import agent
from dotenv import load_dotenv
load_dotenv()
```

```python
from crewai import LLM
llm = LLM(
    model="gemini-2.0-flash",
    temperature=0.1
)
llm.call("Explain me ai agents")
```

```python
from crewai import Agent, Task , Crew
email_assitant = Agent(
    role="Email assistant agent",
    goal="Improve email communication, make it more professional",
    backstory="You are a communication expert. You have to write emails that are professional and efficient.",
    verbose=True,
    llm=llm
)

original_email = """Hello Sir,
this is to inform you that I have completed the tasks we discussed over call. Need your approval for the same."""

email_task = Task(
    description=f"""Write a more professional version of the following email with abbrevations expanded.
    Original email:
    '''{original_email}'''""",
    expected_output="""A professional and more efficient version of the email:""",
    llm=llm,
    agent=email_assitant
)

crew = Crew(
    agents=[email_assitant],
    tasks=[email_task],
    verbose=True
)

result = crew.kickoff()
print(result)
```
